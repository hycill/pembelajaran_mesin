```python
df = pd.read_csv('/content/movies.csv')

# Membuat bar chart
plt.figure(figsize=(40, 12))
colors = ['#FAEBD7' for _ in range(len(df))]
colors[-20] = '#D2691E'
plt.bar(df['movie_name'], df['votes'], color=colors)  # Memperbaiki nama kolom jika perlu
plt.title('Top 100 Movies')
plt.xlabel('Name of Movies')
plt.ylabel('Total Votes of Movie')
plt.xticks(rotation=45, ha='right', fontsize=14)
plt.yticks(range(0,101,20))
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi data di atas merupakan visualisasi dari Corruption Perceptions Index, dapat dilihat bahwa sumbu Y menunjukkan total votes of the movie, sedangkan sumbu X menunjukkan Name of Movies. Skala Corruption Perceptions Index (CPI) berkisar dari 0 (korupsi tinggi) hingga 100 (korupsi rendah).

Pada gambar, terlihat bahwa movie The Matrix menempati peringkat pertama dengan skor votes CPI 1,916,083 , sedangkan Movie The GodFather berada di skor votes 1,860,471. Hal ini menunjukkan bahwa The Matrix  merupakan movie dengan tingkat penayangan terbanyak, sedangkan The Godfather memiliki tingkat penayangan yang sangat rendah.
