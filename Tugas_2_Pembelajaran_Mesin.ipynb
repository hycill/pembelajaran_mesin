```python
df = pd.read_csv('/content/2022worldcup.csv')

# Membuat bar chart
plt.figure(figsize=(20,8))
colors = ['#FBAED2' for _ in range (len(df))]
colors [-29] = '#FE5BAC'
plt.bar(df['Squad'], df['# Pl'], color=colors)  # Memperbaiki nama kolom jika perlu
plt.title('Top Player World cup')
plt.xlabel('Negara')
plt.ylabel('Player')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(range(0,101,20))
plt.tight_layout()
plt.show()
```
